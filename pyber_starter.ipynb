Noticable trends:

#Urban cities have the lowest ceiling on average fare. In Urban areas, trips are more likely to be shorter but happen at a higher frequency. The opposite happens to Rural cities: fewer rides that rack up, on average, a higher price per ride. Suburban cities are between the two extremes of Rural and Urban. Suburban have a lower ride count than Urban while having a higher average price per ride.

#Urban areas included in this data set are bunched closer than the other two city types. This means that drivers would not greatly benefit from changing which Urban city they find their riders. Driving outside an Urban city and moving to the Suburbs does not implicitly mean a higher wage for the driver however because there is a larger spread in Suburban areas for total rides needed on average.

#Urban drivers are the only type to have an inverse relationship between the percentage of drivers to riders. Rural and Suburban drivers have the advantage of having a smaller pool of drivers to get a larger portion of the company's total rides. This means that these drivers have a higher chance to obtain more individual rides than their counterparts in Urban settings. This relationship does not correlate to pure profit for individual drivers, however, since we do not have the data for average trip length, average cost of gasoline, average miles per gallon each driver has, etc. for new drivers to plan on where they would succeed most. For this reason, I would suggest an increase in exposure to Suburban areas as they have a higher price per ride, only really missing a higher demand for rides unlike Urban areas.



```python
# %matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
taxi_df = pd.merge(ride_df, city_df, on="city")

# Display the data table for preview
taxi_df.head()
```

```python
#Manipulate Dataframe for plotting
new_taxi_df = taxi_df.groupby(["city"])


#Calculate total number of rides
total_rides = new_taxi_df['city'].count()


#Calculate average fare per city
average_fare = new_taxi_df['fare'].mean()

#Calculate total drivers per city
total_driver = new_taxi_df['driver_count'].mean()


#Calculate city types
city_type = city_df.set_index('city')['type']



#Creating a new Dataframe for plotting
plt_df = pd.DataFrame({"Average Fare($)":average_fare,
                       "Total Number of Rides (Per City)":total_rides,
                       "Driver Count per City":total_driver,
                       "City Type": city_type})

#Urban
urban = plt_df.loc[plt_df["City Type"]== "Urban"]

#suburban
suburban = plt_df.loc[plt_df["City Type"]== "Suburban"]
suburban.head()

#rural
rural = plt_df.loc[plt_df["City Type"]== "Rural"]


```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
x_urban = urban["Total Number of Rides (Per City)"]
y_urban = urban["Average Fare($)"]
s_urban = urban["Driver Count per City"]

x_rural = rural["Total Number of Rides (Per City)"]
y_rural = rural["Average Fare($)"]
s_rural = rural["Driver Count per City"]

x_suburban = suburban["Total Number of Rides (Per City)"]
y_suburban = suburban["Average Fare($)"]
s_suburban = suburban["Driver Count per City"]

# Build the scatter plots for each city types
plt.scatter(x_urban, y_urban, s= s_urban*10, facecolors= "lightcoral", edgecolors = "black", label = "Urban", alpha = 0.8, linewidth = 1.5)

plt.scatter(x_rural, y_rural, s= s_rural*10, facecolors= "gold", edgecolors = "black",label = "Rural", alpha = 0.8, linewidth = 1.5)

plt.scatter(x_suburban,y_suburban, s= s_suburban*10, facecolors= "lightskyblue", edgecolors = "black",label = "Suburban", alpha = 0.8, linewidth = 1.5 )

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")

plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare($)")

plt.grid()

# Create a legend
lgnd = plt.legend(loc='upper right',frameon = True, title='City Type',markerscale =0.8)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(42,35, 'Note:Circle size correlates with driver count per city',fontsize= '10')


# Save Figure
plt.savefig("../Images/PyberRideSharing.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fare_df = taxi_df.groupby(["type"])

#Calculate Total Fare
total_fare = fare_df["fare"].sum()

# Build Pie Chart
colors = ["lightskyblue","gold","lightcoral"]
explode = [0,0,0.1]
plt.pie(total_fare, explode=explode, labels=total_fare.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

#Incorporate the other graph properties
plt.title("%of Total Fares by City Type")
plt.tight_layout()

# Save Figure
plt.savefig("../Images/Fare_City_Type.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

total_ride = fare_df["ride_id"].count()

# Build Pie Chart
plt.pie(total_ride, explode=explode, labels=total_ride.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

#Incorporate the other graph properties
plt.title("%of Total Rides by City Type")
plt.tight_layout()

# Save Figure
plt.savefig("../Images/Ride_City_Type.png")
```

```python
# Show Figure
plt.show()
```
